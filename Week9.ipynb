```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img_dilasi = cv2.imread('/content/drive/MyDrive/Semester 6/PCVK_MM/Images/j.png')
kernel = np.ones((5,5),np.uint8)
dilasi = cv2.dilate(img_dilasi,kernel,iterations = 1)
 
plt.subplot(131),plt.imshow(img_dilasi)
plt.title('Citra Awal'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(dilasi)
plt.title('Citra Hasil Dilasi'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
img_erosi = cv2.imread('/content/drive/MyDrive/Semester 6/PCVK_MM/Images/j.png')
kernel = np.ones((5,5),np.uint8)
erosi = cv2.erode(img_erosi,kernel,iterations = 1)
 
plt.subplot(131),plt.imshow(img_erosi)
plt.title('Citra Awal'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(erosi)
plt.title('Citra Hasil Erosi'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
img_dilasi2 = cv2.imread('/content/drive/MyDrive/Semester 6/PCVK_MM/Images/j.png')
kernel = np.ones((5,5),np.uint8)
lib_dilasi = cv2.morphologyEx(img_dilasi2,cv2.MORPH_DILATE, kernel)

plt.subplot(131),plt.imshow(img_dilasi2)
plt.title('Citra Awal'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(lib_dilasi)
plt.title('Library Morph Dilasi'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
img_erosi2 = cv2.imread('/content/drive/MyDrive/Semester 6/PCVK_MM/Images/j.png')
lib_erosi = cv2.morphologyEx(img_erosi2, cv2.MORPH_ERODE, kernel)
 
plt.subplot(131),plt.imshow(img_erosi2)
plt.title('Citra Awal'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(lib_erosi)
plt.title('Library Morph Erosi'), plt.xticks([]), plt.yticks([])

plt.show()
```
