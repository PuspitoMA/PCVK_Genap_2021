```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
import math
import os
import glob
```

```python
img = cv.imread('/content/drive/MyDrive/Semester 6/PCVK_MM/Images/lena.jpg')
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
horizontal, vertical, depth = np.shape(img)
n = np.arange(256)

r = [0] * 256
g = [0] * 256
b = [0] * 256

for x in range(horizontal):
  for y in range(vertical):
    r[img[x][y][0]] += 1
    g[img[x][y][1]] += 1
    b[img[x][y][2]] += 1

fig, axs = plt.subplots(1, 3, figsize=[20,5], sharex=True, sharey=True)
fig.suptitle('Histogram RGB Plot')
fig.text(0.05, 0.5, 'Jumlah Kemunculan', va='center', rotation='vertical')
fig.text(0.5, 0.04, 'Intensitas Warna', ha='center')

axs[0].bar(n, r, color='red')
axs[1].bar(n, g, color='green')
axs[2].bar(n, b, color='blue')
```

```python
r = ("r","g","b")
channel = (0, 1, 2)

plt.xlim([0, 256])
for channels, c in zip(channel, r):
  histogram, bin_edges = np.histogram(
      img[:, :, channels], bins=256, range =(0,256)
  )
  plt.plot(bin_edges[0:-1], histogram, color=c)

  plt.xlabel("Intensitas Warna")
  plt.ylabel("Jumlah kemunculan")

  plt.show()
```
